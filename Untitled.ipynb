```r
inequality_df %>%
    filter(`Region` == "London") %>%
    select(`2011 Census Ward code`,
           `2011 Census Ward name`,
           `Local authority code`,
           `Local authority name`,
          `LE1,2 (years)`)
```

```r
codes_in_london <- inequality_df %>%
    filter(`Region` == "London") %>%
    pull(`2011 Census Ward code`)
```

To get a geojson with only London polygons:
1. Load census_wards_supergeneralised.geojson into QGIS
2. Copy and paste the following into the filter
3. Save as a new geojson, taking care to save as EPSG 4326 projection.

```r
cat(paste0('"cmwd11cd" IN (', "'",paste0(codes_in_london, sep="", collapse="','"),"')"))
```
